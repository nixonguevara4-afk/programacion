```python
import numpy as np
import pandas as pd

np.random.seed(123)
```

```python
datos = np.random.normal(loc = 4.0, scale = 0.5, size = 50)
#Simula 50 valores de la normal con media 4 y desviacion estandar 0.5


df = pd.DataFrame({"nota": datos})
```

```python
df.describe()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Asegurar estilo bonito
sns.set(style="whitegrid", context="talk")

# Extraer datos (sin NaN)
data = df["nota"].dropna()

# Parámetros de la normal ajustada
mu = data.mean()
sigma = data.std()

# Crear rango para la curva
x = np.linspace(data.min(), data.max(), 1000)

# Crear figura
plt.figure(figsize=(10,6))

# Histograma (normalizado)
sns.histplot(data, bins=20, stat="density", color="#69b3a2", alpha=0.6, edgecolor="black", label="Histograma")

# FDP normal
plt.plot(x, norm.pdf(x, mu, sigma), color="#d62828", lw=2.5, label=f"Normal ($\\mu$={mu:.2f}, $\\sigma$={sigma:.2f})")

# KDE (kernel density estimate)
sns.kdeplot(data, color="#003049", lw=2.5, label="KDE (Kernel)")

# Detalles estéticos
plt.title("Distribución de la variable 'nota'", fontsize=16, weight="bold")
plt.xlabel("Nota")
plt.ylabel("Densidad")
plt.legend(frameon=True)
plt.tight_layout()

# Mostrar
plt.show()
```
