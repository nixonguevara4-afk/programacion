```python
data = {"nombre": ["Nixon", "Mariana", "Karen", None],
        "edad": [20, np.nan, 23, 30],
        "fruta": ["Guanabana", "Pera", None, "Banana"],
        "pais": [None, "Colombia", "Japón", "España"]}
data
```

```python
data1 = pd.DataFrame(data)
data1
```

```python
data1.isnull() #muestra los valores faltantes
```

```python
data1.isnull().sum() #muestra cuantos valores faltan en cada columna
```

```python
data2 = data1.dropna() #elimina las filas con valores faltantes, en este caso todos
data2
```

```python
data1.dropna(subset=["fruta"]) #elimina las filas con valores faltantes en la columna fruta
```

```python
data1["edad"] = data1["edad"].fillna(data1["edad"].mean()) #rellena los valores faltantes con la media de la columna
data1["nombre"] = data1["nombre"].fillna("Sin nombre")
data1
```

```python
data1["pais"] = data1["pais"].replace("Japón", "JAPON") #remplaza los elementos
data1
```

```python
data3 = data1["pais"].drop_duplicates() #borra los duplicados en este caso no hay duplicados
data3
```

```python
dfn = pd.DataFrame({"ID": [1, 2, 3, 4],
                    "nombre": ["Nixon", "LAURA", "Maria", "Karen"],
                    "edad": [20, 25, 23, 24],
                    "salario": [1000, 2000, 3000, 4000,],
                    "ciudad": ["Bucaramanga", " Medellin", "Bogota", "Cali"]})
dfn
```

```python
dfn["salario_anual"] = dfn["salario"] * 12
dfn["edad_doble"] = dfn["edad"].apply(lambda x: x*2)
dfn
```

```python
dfn = dfn.rename(columns = {"edad": "age"})
dfn
```

```python
dfn["nombre"] = dfn["nombre"].str.lower()
dfn["nombre"] = dfn["nombre"].str.strip()
dfn["nombre"] = dfn["nombre"].str.replace("Nixon", "nixon")
```
